```python
import numpy as np
import tensorflow as tf 
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
```

```python
##loading the dataset
max_features = 10000  # vocabulary size
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=max_features)
```

```python
(X_train, y_train), (X_test, y_test) 
```

```python
X_train[0],y_train[0]
```

```python
##inspect a sample review and its label
sample_review = X_train[0]
sample_label = y_train[0]

print(f"Sample review (as indices): {sample_review}") 
print(f"Sample label: {sample_label}")  
```

```python
##mapping of words index back to words 
word_index = imdb.get_word_index()
word_index 
reverse_word_index= {value: key for (key, value) in word_index.items()} 
reverse_word_index  
```

```python
decoded_review = ' '.join([reverse_word_index.get(i - 3, '?') for i in sample_review])  
```

```python
from tensorflow.keras.preprocessing import sequence

max_input=500 

X_train = sequence.pad_sequences(X_train, maxlen=max_input)
X_test = sequence.pad_sequences(X_test, maxlen=max_input)
X_train
```

```python
##training simple RNN model 
model=Sequential()
model.add(Embedding(input_dim=max_features, output_dim=128, input_length=max_input))
model.add(SimpleRNN(units=128, activation='tanh'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
##create an instance of early stopping  
from tensorflow.keras.callbacks import EarlyStopping
earlystopping=EarlyStopping(monitor='val_loss', patience=5,restore_best_weights=True)
earlystopping
```

```python
##train the model with early stopping
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history=model.fit(
        X_train,y_train,epochs=10,batch_size=32,
        validation_split=0.2,
        callbacks=[earlystopping]
        )

    
```
